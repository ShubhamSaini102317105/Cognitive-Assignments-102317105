Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data, clean and process it, perform statistical analysis, and generate visual insights to help the business make informed decisions.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(102317105)
sales_data=np.random.randint(1000,5001,size=(12,4))
months=['Jan','Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec']
df=pd.DataFrame(sales_data,columns=['Electronics','Clothing','Home & Kitchen','Sports'],index=months)

print(df.head())
print(df.describe())
print(df.sum(axis=0))
sum_month=list(df.sum(axis=1))
print(sum_month)
print(df.diff().mean())
df['Total Sales']=df.sum(axis=1)
growth_rate=[None]
for i in range(1,len(df)):
  prev=df['Total Sales'].iloc[i-1]
  curr=df['Total Sales'].iloc[i]
  change=((curr-prev)/prev)*100
  growth_rate.append(change)
df['Growth Rate']=growth_rate

roll_number=102317105
if(roll_number%2==0):
  df['Electronics']=df['Electronics']-df['Electronics']*0.1
else:
  df['Clothing']=df['Clothing']-df['Clothing']*0.15

plt.plot(df.index, df['Electronics'], marker='o', label="Electronics")
plt.plot(df.index, df['Clothing'], marker='o', label="Clothing")
plt.plot(df.index, df['Home & Kitchen'], marker='o', label="Home & Kitchen")
plt.plot(df.index, df['Sports'], marker='o', label="Sports")
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()

df.boxplot(column=["Electronics", "Clothing", "Home & Kitchen", "Sports"])
plt.title("Sales Distribution per Category")
plt.grid()
plt.show()
```

Q2.For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])

i. Find element-wise absolute value

ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row.

iii. Mean, Median and Standard Deviation of flatt
ened array, of each column, and each row

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

arr=np.array([[1,-2,3],[-4,5,-6]])
print("Absolute values are:-")
print(np.abs(arr))
print("Percentiles of flattened, column-wise and row-wise are:-")
print(np.percentile(arr,[25,50,75]))
print(np.percentile(arr,[25,50,75],axis=0))
print(np.percentile(arr,[25,50,75],axis=1))
print("Mean of flattened,column-wise and row-wise array are:-")
print(np.mean(arr))
print(np.mean(arr,axis=0))
print(np.mean(arr,axis=1))
print("Median of flattened,column-wise and row-wise array are:-")
print(np.median(arr))
print(np.median(arr,axis=0))
print(np.median(arr,axis=1))
print("Standard Deviation of flattened,column-wise and row-wise array are:-")
print(np.std(arr))
print(np.std(arr,axis=0))
print(np.std(arr,axis=1))
```

Q3. For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).

i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np
a=np.array([-1.8,-1.6,-0.5,0.5,1.6,1.8,3])
print("Floor:", np.floor(a))
print("Ceiling:", np.ceil(a))
print("Truncated:", np.trunc(a))
print("Rounded:", np.round(a))
```
